1.   Haz una tabla de frecuencia de 'yearOfStart' y visualízala por año.
- Ordenar por año.
- Dibuja un gráfico de líneas desde 1997 hasta 2014. <= Pista: plt.xlim()

**Respuesta:**

````python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline

# Lea los datos.
# Vaya al directorio donde se encuentra el archivo de los datos.
#os.chdir(r'~~') # Reemplace con la dirección del archivo

df = pd.read_csv('/content/drive/MyDrive/semana 13/sem11_quiz_capítulo3-DATASET/data_coffeeshop.csv', header='infer', na_values=[' '])
df.shape
df.head(5)

DF = df["Year of Start"].value_counts().sort_values()

print(DF.head())


plt.plot(DF.index, DF.values, marker="o")
plt.xlabel("Años")
plt.ylabel("Frecuencia")
plt.title("Grafico desde 1997 - 2014")
plt.xlim(1997, 2014)
plt.grid(True)
plt.show()
````
