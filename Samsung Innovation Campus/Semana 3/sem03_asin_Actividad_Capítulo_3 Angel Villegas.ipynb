# Actividad: Creacion de funciones.


## Instrucciones

1. Defina en forma breve el concepto de función.

2. Crear una función para realizar la suma de los números impares en el rango de 0 a 100, dentro de la función imprima los valores sumados y al final retorne el resultado de la sumatoria.


```python
# Espacio para desarrollar el código requerido, recuerde agregar algunos

# Una función es un fragmento de código, el cual se define para la realización de una operación o resolución
#de un problema en particular, y que puedo "llamar" para reutilizarla en mi código principal según lo necesite,
#se hace con la finalidad de no sobrecargar el flujo principal, y adaptar a diferentes datos una misma lógica.
# Las funciones aceptan parametros de entrada y devolver el resultado de sus operaciones internas.

# comentarios breves sobre su funcionamiento.

def suma_impares():
  num = list(range(1,100,2))
  suma = sum(num)                                                    #sumar cada uno de estos valores entre sí y almacenar en dicha variable
  print("la suma de los números impares del 0 al 100 es: ", suma)    #una vez llamada la función imprime un mensaje con la operación realizada
  return suma                                                        #devuelve el valor de la operación para ser almacenado fuera de la función en una variable


resultado = suma_impares()                                           #llamada a la función, se ejecuta e imprime el mensaje, además de almacenar el valor que retorna
print(resultado)


```

1. Revise documentación relacionada a los procesos de encriptación, explique brevemente como funcionan.
2. Ambas funciones deben tener como entrada un string y una ¨Clave", la ¨Clave¨ será usada para codificar y decodificar.
3. El desarrollo de está actividad esta abierto a multiples soluciones, por favor sea original en su respuesta.


```python
#El mismo es un proceso mediante el cual, una información que es ingresada o recibida a través del programa,
#sea modificada de tal manera que sea ilegible o indescifrable, para aquellos quienes no tenga un acceso autorizado
#a la misma, por lo que el usuario autorizado dispondrá de una autenticación o clave a ser validada por el programa
#para poder descisfrar y mostrarsela.

# Espacio para desarrollar la actividad, si tienes alguna duda o comentario adicional
# lo puedes dejar planteado acá.

def codificar(texto, clave):
    codificado = ""
    for caracter in texto:
        codificado += chr((ord(caracter) + clave) % 256)
    return codificado

def decodificar(texto_codificado, clave):
    decodificado = ""
    for caracter in texto_codificado:
        decodificado += chr((ord(caracter) - clave) % 256)
    return decodificado


usuario = input("Ingrese una cadena de texto: ")
clave1 = int(input("ingrese una clave: "))

texto_codificado1 = codificar(usuario, clave1)
print(f"El texto que acabo de ingresar se codifica como: ",texto_codificado1)

texto_descodificado1 = decodificar(texto_codificado1, clave1)
print(f"El texto descodificado es: ",texto_descodificado1)

```

```python

# Celda para desarrollar
import os                                         #se importa el modulo os que nos permite interactuar con el sistema operativo

def listar_archivos(ruta):                        #se define una función con el parametro ruta el cual representa la ruta del directorio que se desea explorar.
    if os.path.isdir(ruta):                       # verifica si la ruta proporcionada es un directorio. Si es así, se ejecutará el if
      for elemento in os.listdir(ruta):           # bucle a iterar en cada elemento contenido
        ruta_completa = os.path.join(ruta, elemento)  # devuelve una lista de todos los elementos (archivos y carpetas) que se encuentran en el directorio especificado
        if os.path.isdir(ruta_completa):              # verifica si el elemento actual (la ruta completa) es un directorio.
          listar_archivos(ruta_completa)                # si es un directorio, se llama recursivamente a la función
        else:
          print(f"Archivo encontrado: {ruta_completa}")
ruta_principal = "Carpeta_principal"
listar_archivos("Carpeta_principal")


```

# Actividad: Uso de la funcion Lambda.

## Instrucciones

1. Escriba una función regular donde se realice una operación.
2. Escriba la misma función donde use la expresión lambda.
3. Explica brevemente las ventajas de usar la función Lambda.


```python
# Espacio para desarrollar el código siguiendo las instrucciones.

def exponente(base, exponente):
    operacion = base ** exponente
    return operacion

print("el resultado de la operación exponente es:", exponente(5,10))


exponente2 = lambda base, exponente: base ** exponente

print("el resultado de la operación exponente es:", exponente2(5, 10))

# Puede decirse que la ventaja más evidente es que ahorramos líneas de codigo, para terminar ejecutando en este caso la misma función,
# la unica condición es que debe ser una operación sencilla, de tal manera de simplificarse en una sóla línea de codigo.
# Si nuestro programa es muy extenso nos conviene recurrir a este tipo de funciones, para hacerlo más legible y depurable a posibles errores.
```
