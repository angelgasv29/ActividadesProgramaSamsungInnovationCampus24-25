## Q 08-03

### ¿Qué sucede si el error de verificación aumenta constantemente cuando graficamos el error de verificación para cada época utilizando el método de descenso de pendiente por lotes? Además, ¿cómo se puede solucionar este problema?

Si al graficar el error de verificación de un modelo de aprendizaje automático utilizando el método de descenso de pendiente por lotes vemos que el error aumenta constantemente, esto puede ser un signo de varios problemas. Uno de los más comunes es el sobreajuste, donde el modelo aprende demasiado bien los datos de entrenamiento pero no puede generalizar a los datos de validación. Otra posible causa es que la tasa de aprendizaje sea demasiado alta, lo que puede hacer que el modelo oscile y no converja adecuadamente.

Para solucionar este problema, existen varias acciones a las que pueden recurrirse, una opción es usar técnicas de regularización, como L1 o L2, que ayudan a controlar la complejidad del modelo. También se puede ajustar la tasa de aprendizaje, bajándola o utilizando algoritmos que se adapten durante el entrenamiento. Otra alternativa es simplificar el modelo, usando menos nodos o capas, si es posible, aumentar el conjunto de datos o aplicar técnicas de aumento de datos también puede ser útil, con estas medidas, se puede mejorar el rendimiento del modelo y evitar problemas con el error de verificación.

## Q 08-05

### Aplique la función softmax cuando la salida de la red neuronal sea $(0.4,2.0,0.001,0.32)^T$ y escriba el resultado.

Responder aquí

```python
import numpy as np

# Esta es la función softmax

def softmax(z):
  exp_z = np.exp(z - np.max(z))
  return exp_z / np.sum(exp_z)


z = np.array([0.4, 2.0, 0.001, 0.32])

resultado = softmax(z)

print("resultado de softmax: ", resultado)
```
