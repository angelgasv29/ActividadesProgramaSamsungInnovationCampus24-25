## Q 07-01

### ¿Cuál de las siguientes afirmaciones sobre la minería de textos es incorrecta?
1. Puedes conocer la reacción ante un grupo específico.
2. Se puede obtener información sobre oportunidades de ventas.
3. Se pueden establecer estrategias competitivas monitoreando otras marcas.
4. El mismo método se puede aplicar a varios idiomas.

La afirmación que es incorrecta es la opción 4. El mismo metodo se puede aplicar a varios idiomas.

La razón es que todos los idiomas existen diferencias sustanciales en la gramatica, vocabulario y estructura lingüistica, por ello, no en todos esos idiomas podremos aplicar y obtener resultados igual de efectivos para un mismo método.

## Q 07-02

### ¿Cuál de las siguientes afirmaciones no es cierta sobre el uso de las redes sociales?
1. Es posible saber de cuántos grupos está compuesta la red.
2. Es posible conocer clientes influyentes.
3. Es posible ver el cambio a lo largo del tiempo.
4. Es posible saber si los clientes se irán la próxima vez.

Podría decirse que la afirmación que no es del todo cierta es la 4. Es posible saber si los clientes se irán la próxima vez.

En tal caso podemos estimar la probabilidad de que ese suceso de irse se dé o no, pero nunca determinar o probar con tanta exactitud esta afirmación para todos los casos, en donde pueden influir diversos factores más allá de medible o evidente.

## Q 07-03

### ¿Qué significa el término separar una raíz de una palabra cuya forma ha sido modificada para extraerla? ¿Cuál es la palabra que es objeto de análisis de morfemas?

A dicho método en el contexto del Procesamiento del lenguaje natural (PLN) se le conoce como stemming, que se dice que reduce la palabra a su raíz, aunque existen casos que la palabra reducida resultante puede ser una palabra no válida, por lo que dependiendo de esto, es que puede resultar efectivo o no éste método.

## Q 07-04

### ¿Qué significa el término un conjunto de materiales que pueden mostrar los aspectos esenciales del lenguaje como material de investigación requerido en cada campo de la investigación del lenguaje?

Dentro del PLN el termino "conjunto de materiales que pueden mostrar los aspectos esenciales del lenguaje ..." Puede estarse refiriendo al corpus, informalmente se conoce como una bolsa en donde se vacia una estructura desglosada de textos, que se utilizan para estudiar las caracteristicas del lenguaje, como gramática, el vocabulario y el uso del lenguaje en diferentes contextos.

## Q 07-05

### Explique sobre TF (Frecuencia de términos) y escriba cómo calcular TF (fórmula).

Nuevamente en el contexto del PLN, se trata de una medida, que nos permite a través de la frecuencia conocer la importancia de una palabra en un "documento" hablando de forma generica sea a un conjunto de frases, oraciones, texto o documentos propiamente dicho, mientras más se repita una frase más relevante puede ser esa palabra para el contexto de ese documento.

Es importante mencionar que para este mismo principio se aplica para "penalizar" a aquellas palabras que se repitan excesivamente, restrigiendo un poco la relevancia de su aparición y devolverle la importancia a las palabras que más se repiten que le anteceden a estas o incluso darle importancia a aquellas palabras que menos se repitan como las más relevantes.

TF(t,d) = Número de veces que aparece el termino t en el documento d/ Número total de términos en el documento d.
