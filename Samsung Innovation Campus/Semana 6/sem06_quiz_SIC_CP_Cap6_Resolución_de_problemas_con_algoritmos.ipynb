## Q 30-01

Este es el problema del intercambio de monedas.

Tienes una moneda de 400 wones.

Escriba el resultado de cómo el algoritmo `coin_change()` determina el cambio de la moneda de 800 wones.
```
coins  = [500, 400, 100, 50, 10]
amount = int(input("Ingrese la cantidad: "))
changes= coin_change(coins, amount)
print(changes, len(changes))

Ingrese la cantidad: 800
```

````python
def coin_change(monedas, monto):
	cambio = []
	mayor = 0
	while monto > 0:
		if monto % monedas[mayor] == 0:
			cambio.append(monedas[mayor])
			monto -= cambio[cambio.index(monedas[mayor])]
		elif monedas[mayor] > monto:
			mayor += 1
		else:
			mayor += 1
			cambio.append(monedas[mayor])
			monto -= cambio[cambio.index(monedas[mayor])]
	return cambio

monedas = [500, 400, 100, 50, 10]
cantidad = int(input("Introduzca monto: "))
cambios = coin_change(monedas, cantidad)
print(cambios, len(cambios))
````

## Q 31-01

Tienes 8 monedas idénticas numeradas del 1 al 8. De estas, solo una moneda es más pesada que la otra.

Suponiendo que puede pesar monedas con una balanza equilibrada, diseñe un algoritmo que encuentre la moneda más pesada.

En este momento, ¿cuántas veces se debe usar al menos la balanza?

````python
import random

def encontrar_moneda_mayor(monedas, menor, mayor):
    global pesadas
    if menor == mayor:
        return menor

    medio = (menor + mayor) // 2

    peso_izquierda = sum(monedas[menor:medio + 1])
    peso_derecha = sum(monedas[medio + 1:mayor + 1])

    pesadas += 1

    if peso_izquierda > peso_derecha:
        return encontrar_moneda_mayor(monedas, menor, medio)
    else:
        return encontrar_moneda_mayor(monedas, medio + 1, mayor)

monedas = [1] * 8
indice_aleatorio = random.randint(0, len(monedas) - 1)
monedas[indice_aleatorio] = 2

print("Monedas:", monedas)

pesadas = 0

indice_mayor = encontrar_moneda_mayor(monedas, 0, len(monedas) - 1)
print("El índice de la moneda mayor es:", indice_mayor)
print("Número de pesadas realizadas:", pesadas)

````

## Q 31-02

Hay nueve monedas idénticas numeradas del 1 al 9 en el problema anterior.

Diseñe un algoritmo para encontrar una moneda pesada en este caso.

En este momento, ¿cuántas veces se debe usar al menos la balanza?

````python
import random

def encontrar_moneda_pesada(monedas, inicio, fin):
    global pesadas
    if inicio == fin:
        return inicio

    medio1 = inicio + (fin - inicio) // 3
    medio2 = inicio + 2 * (fin - inicio) // 3

    peso1 = sum(monedas[inicio:medio1 + 1])
    peso2 = sum(monedas[medio1 + 1:medio2 + 1])
    pesadas += 1

    if peso1 > peso2:
        return encontrar_moneda_pesada(monedas, inicio, medio1)
    elif peso1 < peso2:
        return encontrar_moneda_pesada(monedas, medio1 + 1, medio2)
    else:
        return encontrar_moneda_pesada(monedas, medio2 + 1, fin)


monedas = [1] * 9
indice_aleatorio = random.randint(0, len(monedas) - 1)
monedas[indice_aleatorio] = 2

print("Monedas:", monedas)

pesadas = 0

indice_pesada = encontrar_moneda_pesada(monedas, 0, len(monedas) - 1)
print("El índice de la moneda pesada es:", indice_pesada)
print("Número de pesadas realizadas:", pesadas)
````

## Q 32-02

Analice el resultado de la ejecución del siguiente código y compárelo con el rendimiento de las funciones `bin1()` y `bin2()`.
```
def bin3(n,k):
    B = [0] * (n + 1)
    for i in range(n + 1):
        for j in range(n, -1, -1):
            if j==0 or j==i:
                B[j] = 1
            else:
                B[j] = B[j] B[j-1]
    return B[k]


for i in range(10):
    for j in range(i+1):
        print(bin3(i,j),end=" ")
    print()
```

La función bin3 ofrece un enfoque balanceado y eficiente para calcular coeficientes binomiales, especialmente para valores relativamente pequeños de
n. En comparación, bin1 es ineficiente y bin2 es una buena alternativa si se necesita optimización adicional. Si el rendimiento es crítico y el espacio es limitado, bin3 es una opción sólida, aunque bin2 podría ser preferida si se busca la máxima eficiencia en tiempo.

````python
def bin3(n,k):
    B = [0] * (n + 1)                 #O(1)
    for i in range(n + 1):            #O(n)
        for j in range(n, -1, -1):    #O(n)
            if j==0 or j==i:
                B[j] = 1              #O(1)
            else:
                B[j] = B[j] + B[j-1]  #O(1)
    return B[k]                       #Dominancia de terminos final O(n**2)


for i in range(10):                  #O(n)
    for j in range(i+1):             #O(n)
        print(bin3(i,j),end=" ")     #O(1)
    print()                          #O(1)
                                     #Dominancia de terminos final O(n**2)
````

## Q 33-01

Si reemplaza cada letra con un número en la siguiente letra, el número correspondiente a las cinco palabras son todos cuadrados.

Para cada palabra, la suma de cada dígito también es un número cuadrado.

Encuentra el número representado por cada letra.

**Código de referencia:**

````python
from math import sqrt, floor
import time


def is_square_digitsum(n):
    s = 0
    while n > 0:
        s += n % 10
        n //= 10
    if sqrt(s) == int(sqrt(s)):
        return True
    return False


def find_all_squares():
    sqrs = [[] for _ in range(5)]
    for i in range(1, floor (sqrt(10 ** 5)) + 1):
        n = i * i
        if not is_square_digitsum(n):
            continue
        s = str(n)
        if len(s) == 3 and s[1] != s[2]:
            continue
        if len(s) == 5 and s[2] != s[3]:
            continue
        if len(s) in [4, 5] and len(set(s)) != 4:
            continue
        sqrs[len(s) - 1].append(n)
    return sqrs


def promising (s, n, dic):
    for i in range(len(s)):
        digit = int(str(n)[i])
        for key, value in dic.items():
            if key == s[i] and value != digit:
                return False
            if value == digit and key != s[i]:
                return False
    return True


def solve(words, dic, squares):
    global solved
    if (len(words) == 0):
        solved = dic
    else:
        s = words[0]
        candidates = squares[len(s) - 1]
        for n in candidates:
            if promising(s, n, dic):
                newdic = dic.copy()
                for i in range(len(s)):
                    newdic[s[i]] = int(str(n)[i])
                solve(words [1:], newdic, squares)


def main():
    squares = find_all_squares()
    print(squares)
    words = ['A', 'TO', 'ALL', 'XMAS', 'MERRY']
    dic = {}
    solve (words, dic, squares)
    for word in words:
        print (word, end=": ")
        for c in word:
            print(solved[c], end="")
        print()

start = time.time()
solved = {}
main()
end = time.time()
print("Elapsed Time:", end-start,"seconds")
print(solved)
````
