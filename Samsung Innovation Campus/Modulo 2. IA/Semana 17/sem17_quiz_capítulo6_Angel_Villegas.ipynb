## Q 06-04

### Escribe tres o más ejemplos de algoritmo de clustering

1. k-means el cual divide los datos en "K clusters", donde cada punto pertenece al cluster cuyo centroide es el más cercano.

2. DBSCAN: Agrupa puntos que están densamente conectados, identificando áreas de alta densidad y separando puntos en áreas de baja densidad como ruido.

3. k-medoides: como  técnica de clustering que utiliza medidas de similitud y disimilitud para agrupar los datos, a diferencia del algoritmo K-means que usa puntos aleatorios como centros de los grupos.

Método del codo: Se define la suma de las distancias al cuadrado, entre las observaciones y los centroides más cercanos. "Suma interna de los cuadrados frente a n° de conglomerados, para encontrar el punto de inflexión "Codo".

Criterio de información Bayesiano: se usa una fórmula matemática (el BIC) para evaluar qué tan bien se ajusta el modelo de clustering con diferentes valores de k, seleccionando el número de clusters que minimiza el BIC, lo cual indica el modelo que mejor se adapta a los datos sin ser demasiado complejo.
