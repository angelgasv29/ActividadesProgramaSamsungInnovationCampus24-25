## Q 27-01

¿Cuántas comparaciones se ejecutaron en el siguiente proceso de ordenamiento de burbuja?
```
def bubbleSort(S):
    n = len(S)                         #5
    for i in range(n):                 #0-4
        print(S)                       
        for j in range(n-1):           #0-3
            if S[j] > S[j+1]:
               S[j] , S[j+1] = S[j+1] , S[j]

S = [50, 30, 40, 10, 20]
bubbleSort(S)
print(S)
```
#### Output
```
[50, 30, 40, 10, 20]
[30, 40, 10, 20, 50]
[30, 10, 20, 40, 50]
[10, 20, 30, 40, 50]
[10, 20, 30, 40, 50]
[10, 20, 30, 40, 50]
```

La cantidad de comparaciones que se realizan en total son 20.

El primer bucle define cuantas veces se va a repetir el segundo bucle, el que evalua y compara, si un elemento desde el punto de partida de la lista, es mayor que el de la siguiente posición, y realizar o no el intercambio de izquierda a derecha, asegurando que el elemento mayor en la derecha, y ordenar el menor con respecto a ese elemento, en la izquierda.

Nuevamente comparar ese elemento "mayor" con el de la siguiente poisición derecha, para realizar o no el intercambio, hasta finalizar la iteración del segundo bucle en toda la lista. En la repetición definida por el primer bucle, se vuelven a revisar todos los elementos de la lista, pese a que esten ordenados con respecto al resto y no se puedan realizar intercambios por no cumplirse dicha condición.

## Q 28-01

¿Cuántas veces se ejecuta la función `merge2()` en el siguiente proceso de ordenamiento por fusión?
```
S = [6,2,11,7,5,4,8,16,10,3]
mergeSort2(S, 0, len(S)-1)
print(S)
```

````python
count = 0
def mergesort2(lista, menor, mayor):
    if menor < mayor:
        print(lista)
        mitad = (menor + mayor) // 2
        mergesort2(lista, menor, mitad)
        mergesort2(lista, mitad + 1, mayor)
        merge2(lista, menor, mitad, mayor)

def merge2(lista, menor, mitad, mayor):
    global count
    count += 1
    auxiliar = []
    i, j = menor, mitad + 1
    while i <= mitad and j <= mayor:
        if lista[i] < lista[j]:
            auxiliar.append(lista[i])
            i += 1
        else:
            auxiliar.append(lista[j])
            j += 1

    if i > mitad:
        for k in range(j, mayor + 1):
            auxiliar.append(lista[k])
    else:
        for k in range(i, mitad + 1):
            auxiliar.append(lista[k])

    for k in range(len(auxiliar)):
        lista[menor + k] = auxiliar[k]
    print(count)

S = [6, 2, 11, 7, 5, 4, 8, 16, 10, 3]
resultado = mergesort2(S, 0, len(S) - 1)
print(S)
````

La función merge2() se ejecutará 9 veces durante el proceso de ordenamiento.

## Q 29-01

Dada la lista a continuación, escriba la salida después de ejecutar la función `partition1()`
```
S = [15,10,12,20,25,13,22]
partition1(S, 0, len(S)-1)
print(S)
```
#### Output
```
[15,10,12,20,25,13,22] 0 6 pivot = 15
...
```


````python
def partition1(lista, menor, mayor):
  pivot = lista[menor]
  izquierda, derecha = menor + 1, mayor
  while izquierda < derecha:
    while izquierda <= derecha and lista[izquierda] <= pivot:
      izquierda += 1
    while izquierda <= derecha and lista[derecha] >= pivot:
      derecha -= 1
    if izquierda < derecha:
      lista[izquierda], lista[derecha] = lista[derecha], lista[izquierda]
  pivotpoint = derecha
  lista[menor], lista[pivotpoint] = lista[pivotpoint], lista[menor]

  return pivotpoint

S = [15,10,12,20,25,13,22]
resultado = partition1(S, 0, len(S)-1)

print(f"{S}, pivot = {S[resultado]}")
````
