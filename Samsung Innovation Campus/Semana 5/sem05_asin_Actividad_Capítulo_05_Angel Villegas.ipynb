# Actividad: Ordenamiento de Burbuja


## Instrucciones

```python
# Espacio para desarrollar el código requerido, recuerde agregar algunos
# comentarios breves sobre su funcionamiento.


"""Ordenamiento por burbujas consiste en recorrer una lista de valores numericos, con un bucle anidado, comparando
 e intercambiando "elementos adyacentes" desde el principio hasta el final de la lista, (de dos en dos) comenzando
por el primero y el segundo, después el segundo con el tercero, y así sucesivamente. Para asegurar su ordenamiento,
al principio el segundo bucle debe iterar tantas veces la cantidad de elementos que indique su longitud, indice por indice
hasta que no se realicen más intercambios para ordenar, por lo que en cada vuelta o repetición del primer bucle, un elemento
deberá quedar ordenado con respecto al respecto al resto, "el elemento más grande burbuja" se ubicará en la primera
o ultima posición, es decir, si la lista ordenada es ascendente (de menor o mayor), o descendente (mayor a menor)."""



def bubblesort(lista):
	n = len(lista)
	for i in range(n):                                             #Se define un primer bucle el cual nos va a permitir repetir las operaciones del segundo bucle.
		for j in range(n - 1 - i):                                   #El segundo bucle evalua, ordena e intercambia, los elementos según la condición (de mayor a menor o viceversa).
			if lista[j] > lista[j + 1]:                                #Va desde el rango de los elementos que falten por ordenar según la condición anterior, se delimita el rango,
				lista[j], lista[j+ 1] = lista[j + 1], lista[j]           #para evitar volver a ordenar los que ya lo estén.
	return lista

Lista = [25,30,11,-14,0,1,32,100,22]

print(bubblesort(Lista))


```

```python
# Celda para desarrollar
"""El Quick Sort es un método eficiente de ordenamiento que utiliza el término o técnica de 'divide y venceras'
como primer paso tenemos que seleccionar un pivote, se elige un elemento de una lista como pivote, posteriormente se realiza
una partición, es decir se reorganiza la lista de tal forma que los elementos menores al pivote queden a su izquierda y los
elementos mayores a su derecha. Se llama recursivamente a la función quicksort y para aplicarse exactamente el mismo proceso,
todo esto hasta que las sublistas tengan uno o cero elementos, representando que ya están ordenadas.
"""

def quicksort1(lista, menor, mayor):
  if menor < mayor:
    print(lista)
    pivotpoint = particion(lista, menor, mayor)
    quicksort1(lista, menor, pivotpoint - 1)           #ordena la sublista izquierda llamando recursivamente así mismo
    quicksort1(lista, pivotpoint + 1, mayor)           #ordena la sublista derecha

def particion(lista, menor, mayor):
  pivot = lista[menor]                                 #se elige el primer elemento como pivote
  izquierda, derecha = menor + 1, mayor

  while izquierda <= derecha and lista[izquierda] <= pivot:  #reorganizar la lista en torno al pivote
    izquierda += 1

  while izquierda <= derecha and lista[derecha] >= pivot:
    derecha -= 1

  if izquierda < derecha:                                    #Intercambiar elementos si es necesario
    lista[izquierda], lista[derecha] = lista[derecha], lista[izquierda]

  pivotpoint = derecha                                       #colocar el pivote en su posición final

  lista[menor], lista[pivotpoint] = lista[pivotpoint], lista[menor]

  return pivotpoint                                           #retorna la posición final del pivote.



lista = [0,33,1,10,17,19,22,2]

quicksort1(lista, 0, len(lista) - 1)
print(lista)
```
